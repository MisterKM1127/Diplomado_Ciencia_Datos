Los cuartiles en estadística son medidas que dividen un conjunto de datos ordenados en cuatro partes iguales. Estas medidas son utilizadas para entender la distribución y la dispersión de los datos en un conjunto. Los tres cuartiles principales son:

1. Primer cuartil (Q1): También conocido como el cuartil inferior, es el valor que separa el 25% más bajo de los datos del 75% restante. Es el valor que ocupa el lugar central entre el valor más bajo y la mediana.

2. Segundo cuartil (Q2): Es la mediana del conjunto de datos. Divide el conjunto en dos partes iguales, con el 50% de los datos por encima y el 50% por debajo de este valor.

3. Tercer cuartil (Q3): También llamado el cuartil superior, divide el 75% más bajo de los datos del 25% más alto. Al igual que el primer cuartil, ocupa el lugar central entre la mediana y el valor más alto.

Los cuartiles son especialmente útiles para identificar la dispersión y presencia de datos atípicos o valores extremos en un conjunto de datos. Además, se utilizan frecuentemente para construir diagramas de caja (boxplots), que proporcionan una representación visual de la distribución de los datos y ayudan a detectar valores atípicos en los datos.

### <span style="color:blue">**Formula matematica**</span>

$ Q1 = X_{\left(\lceil n \times 0.25 \rceil\right)} \\
Q2 = \begin{cases} X_{\left(\frac{n+1}{2}\right)} & \text{si } n \text{ es impar} \\ \frac{X_{\left(\frac{n}{2}\right)} + X_{\left(\frac{n}{2}+1\right)}}{2} & \text{si } n \text{ es par} \end{cases} \\
Q3 = X_{\left(\lceil n \times 0.75 \rceil\right)} $
