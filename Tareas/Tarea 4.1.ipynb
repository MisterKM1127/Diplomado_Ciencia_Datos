Escribe una calse con las siguientes pasos y funcionalidades:
1. Consiga el conjunto de datos iris.csv
2. Escriba una clase iterable que tenga la siguiente funcionalidad
3. Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.
4. Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
5. Imprima "error" si los datos se han agotado.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/AprendizajeProfundo/Diplomado-Avanzado/c01c908a0388a58b45cac707cb716438197a2d88/M%C3%B3dulo%2011%20-%20Mapas%20Auto%20Organizados/Datos/iris.csv"
data = pd.read_csv(url)

class IrisDataIterator:
    def __init__(self, data):
        self.dataframe = data
        self.index = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.dataframe):
            raise StopIteration("\nerror")

        rows = self.dataframe.iloc[self.index:self.index+2]
        self.index += 2
        self.count += 1
        return rows

    def get_count(self):
        return self.count

```

Ya construimos la clase, ahora vamos a mostrar lo que se pide

```python
data_iterator = IrisDataIterator(data)

for _ in range(200):
    try:
        data_points = next(data_iterator)
        print(data_points)
    except StopIteration as e:
        print(e)
        break

print("\nTotal de pares de filas recuperados:", data_iterator.get_count())
```
